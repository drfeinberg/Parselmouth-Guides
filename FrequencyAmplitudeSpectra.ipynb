```python
import matplotlib.pyplot as plt
import numpy as np
import parselmouth
```

```python
sound = parselmouth.Sound('03-01-01-01-01-01-01.wav')
spectrum = sound.to_spectrum()
amplitudes = 10 * np.log10(spectrum.values[0])
frequencies = np.array([spectrum.get_frequency_from_bin_number(bin + 1) for bin in range(spectrum.get_number_of_bins())])
plt.plot(frequencies, amplitudes, color='black', linewidth=0.1)
plt.xlim(0, 5500)
plt.ylim(-60, np.nanmax(amplitudes))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude (dB)")
```
