```python
import matplotlib.pyplot as plt
import numpy as np
import parselmouth
```

```python
sound = parselmouth.Sound('03-01-01-01-01-01-01.wav')
spectrum = sound.to_spectrum()
spectrum_values = spectrum.values[0,:] + 1j * spectrum.values[1,:]
power_spectral_density = 10 * np.log10(2 * abs(spectrum_values)**2 * spectrum.dx / 4e-10)
frequencies = np.array([spectrum.get_frequency_from_bin_number(bin + 1) for bin in range(spectrum.get_number_of_bins())])
plt.plot(frequencies, power_spectral_density, color='black', linewidth=0.1)
plt.xlim(-4e-10, 5500)
plt.ylim(0, np.nanmax(power_spectral_density))
plt.xlabel("Frequency bin (Hz)")
plt.ylabel("Amplitude (dB) / Frequency(Hz)")
```
